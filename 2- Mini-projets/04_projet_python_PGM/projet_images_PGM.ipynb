## Fichier PGM

Un fichier PGM (Portable Gray Map) est compos de deux parties : 

* Un entte indiquant les paramtres de l'image
* Une matrice dont chaque lment reprsente le niveau de gris dun pixel de limage.

 Lentte contient 3 lignes. La premire ligne est toujours la mme, il sagit de lidentifiant du fichier PGM : P2 pour une image code en ASCII et P5 pour une image code en binaire. La seconde ligne contient 2 entiers spars par un espace: le nombre de colonnes et le nombre de lignes de pixels de limage.

  



 Enfin, la troisime ligne contient la valeur maximale que peut prendre un pixel (par exemple : 255 pour un codage des niveaux de gris entre 0 et 255). 
 Les lignes qui suivent donnent lintensit de chaque pixel. Ceux-ci sont numrs ligne par ligne et colonne par colonne. Ainsi la premire valeur est lintensit du pixel du coin suprieur gauche, la seconde est lintensit du pixel de la premire ligne, deuxime colonne, etc.

Par exemple, un fichier PGM contenant les informations ci-dessous est lencodage dune image reprsentant une barre oblique blanche sur fond noir (voir Fig. 1 du sujet en pdf).

Quelle est la structure du tuple retourn? O se trouve le tableau

A quoi correspond le triplet (data[1],data[0]),data[2]) ?

```python
# CODE FOURNI AUX APPRENANTS
def readImg(filename):
    with open(filename) as f:
        lines = f.readlines()

    # Ignores commented lines
    for l in list(lines):
        if l[0] == '#':
            lines.remove(l)

    # Makes sure it is ASCII format (P2)
    # leading and trailing whitespace removed
    assert lines[0].strip() == 'P2' 

    # Converts data to a list of integers
    data = []
    for line in lines[1:]:
        data.extend([int(c) for c in line.split()])

    return (np.array(data[3:]),(data[1],data[0]),data[2])
```
