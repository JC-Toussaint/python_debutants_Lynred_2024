## simple plot(x,y)
On désire tracer les valeurs d'un tableau y en fonction de celles d'un tableau x

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact

plt.close('all')
# data
def trace(colori='red', linewidth=2, linestyle='-', marker='*'):
    x = np.linspace(0, 2 * np.pi, 400)
    y = np.sin(x ** 2)
    y2 = np.cos(x)

    plt.plot(x, y, label = 'sin(x²)',color=colori, linewidth=linewidth, linestyle=linestyle, marker = marker)
    plt.plot(x, y2, label = 'cos(x)', color='red', linewidth=linewidth, linestyle=linestyle, marker = marker)
    #plt.grid()
    plt.xlabel('axe x')
    plt.ylabel('axe y')
    plt.title('titre')
    plt.legend()
    
    plt.show()

interact(trace)
```

## Assemblage de figures dans une page
On retrouve une fonction similaire à la fonction subplot de Matlab

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

plt.figure()
plt.subplot(2, 2, 1)
plt.plot(x, y, color='red', linewidth=2.0, linestyle="-")
plt.title('titre1')
plt.grid()

plt.subplot(2, 2, 2)
plt.plot(x, y, color='blue', linewidth=2.0, linestyle="-", label='sin')
plt.title('titre2')
plt.legend(loc='best')

plt.subplot(2, 1, 2)
plt.plot(x, y, color='red', linewidth=2.0, linestyle="-")
plt.title('titre3')
plt.show()
```
